При разработке ML-систем мы регулярно сталкиваемся с 3 видами метрик:

бизнес-метрики,
метрики качества,
лосс-функции (функции потерь).
Важно понимать, какую роль выполняют эти метрики и как они взаимосвязаны между собой. В этой задаче мы разберем этот вопрос подобно.



Бизнес-метрики
Бизнес-метрики - это показатели бизнеса, которые мы хотим улучшить с помощью внедрения ML-модели. Эти метрики напрямую связаны с целями компании. Это могут быть: выручка, конверсия, отток пользователей, средний чек, стоимость доставки и др.

Бизнес привлекает ML-команду для улучшения какой-то бизнес-метрики. Менеджерам не понятны термины RMSE, ROC-AUC, NDCG. Они понимают только термины средний чек, конверсия, выручка. Оценивать качество разработанной ML-системы бизнес будет по изменениям бизнес-метрик. A/B эксперимент будет проверять статистически значимое изменение бизнес-метрики.

Поэтому важно помнить, что все работы от проектирования и разработки до внедрения ML-решения направлены на улучшение целевой бизнес-метрики. И держать в голове, поможет ли то, что мы делаем как-то повлиять на бизнес-метрику.



Метрики качества
Метрики качества позволяют измерить, насколько хорошо модель выполняет свою задачу с технической стороны. Они помогают разработчикам и аналитикам понимать, насколько хорошо модель справляется. Эти метрики варьируются в зависимости от типа задачи машинного обучения.

В задачах классификации часто используют: Precision, Recall, F1-мера, ROC-AUC.

В задачах регрессии регулярно применяют: MSE (Среднеквадратичная ошибка), MAE (Средняя абсолютная ошибка).

В задачах ранжирования используют:  NDCG (Normalized Discounted Cumulative Gain), Precision@K, Recall@K.

Метрики качества  помогают ML-команде оценивать как модель решает свою задачу. Например, насколько качественно классификатор определяет пользователей, кто не продлит подписку в следующем месяце. Но эти метрики не являются конечной целью внедрения ML-системы. Поэтому важно как можно раньше убедиться, что улучшение метрики качества действительно улучшает целевую бизнес-метрику.

Хорошим способом подтвердить такую корреляцию будет организовать эксперимент, в рамках которого сравниваются две версии модели с разным уровнем метрики качества и проверяется действительно ли модель с лучшим качеством  лучше влияет на целевую бизнес-метрику.

Лосс-функции
Лосс-функции (функции потерь) используются для обучения модели. Они показывают оптимизационному алгоритму насколько далеко предсказания модели от истинных значений.

К лосс-функции предъявляются 2 требования:

Она должна быть дифференцируемой. Многие алгоритмы машинного обучения используют метод градиентного спуска, поэтому возможность взять производную от лосс-функции очень важно.
Она должна быть коррелированна с метрикой качества. Уменьшение лосс-функции должно улучшать выбранную метрику качества.

Взаимосвязь метрик
Работая над ML-системами, наша цель - выбрать такую метрику качества, которая поможет улучшить целевую бизнес-метрику.  Затем выбрать такие ML-алгоритм и лосс-функцию, которые лучше всего помогает увеличить нашу метрику качества. Это как выбрать правильный инструмент, чтобы достичь нашей цели.

Если мы все сделали правильно, то когда мы начинаем улучшать лосс-функцию, улучшается и метрика качества. А улучшение метрики качества приводит к росту бизнес-метрики, а значит и к росту бизнеса.

Мы не можем напрямую с помощью какой-то ML-модели, улучшить конверсию. Но мы можем создать ML-модель, которая хорошо предугадывает что понравится пользователю, а это в свою очередь, поможет увеличить конверсию.

Мы не можем напрямую оптимизировать Retention (удержание пользователей) в почтовом сервисе. Но мы можем внедрить анти-фрод ML-модель, которая будет хорошо классифицировать спам, и, фильтруя такой спам, мы увеличим удовлетворенность пользователей нашим сервисом и увеличим Retention.

Помните эти простые принципы, когда подбираете метрику качества для решения бизнес-задачи и когда подбираете лосс-функцию для улучшения метрики качества.
